A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 × 99.

Find the largest palindrome made from the product of two 3-digit numbers.

```python
def ispolindrom(d):
    return str(d)==str(d)[::-1]

found = 0
for i in range(999, 0, -1):
    for j in range(999, 0, -1):
        if ispolindrom(i * j) and (i*j > found): 
            found = i * j
found
```

1^2 + 2^2 + ... + 10^2 = 385

# Problem 9
## Special Pythagorean triplet
A Pythagorean triplet is a set of three natural numbers, a < b < c, for which,

a^2 + b^2 = c^2

For example, 3^2 + 4^2 = 9 + 16 = 25 = 5^2.

There exists exactly one Pythagorean triplet for which a + b + c = 1000.
Find the product abc.

```python
st = time.perf_counter()
for i in range(500):
    for j in range(500):
        for k in range(500):
            if sum([i, j, k]) == 1000:
                if i**2 + j**2 == k**2:
                    print(i * j * k)
time.perf_counter()-st
# Ошибки: i>j>k. Если это учесть в циклах, то будет выполняться быстрее.
# Третий цикл и первое условие можно не выполнять, а сразу подбирать к такой, чтобы подходил под условие
```

```python
st = time.perf_counter()
n = 1000
for i in range(1, n//3):
    for j in range(i, n - i):
        k = n - i - j
        if i**2 + j**2 == k**2:
            print(i*j*k)
time.perf_counter()-st
```

# Problem 10

Find the sum of all the primes below two million.

```python
def primelist(n):
    '''Create list of prime digit'''
    resheto = set(range(2, n+1))
    prime = []
    while resheto:
        prime.append(min(resheto))
        resheto -= set(range(prime[-1], n+1, prime[-1]))
    return prime

st = time.perf_counter()
sum(primelist(200000)), time.perf_counter()-st
```

```python
import itertools
def erat2( ):
    D = {  }
    yield 2
    for q in itertools.islice(itertools.count(3), 0, None, 2):
        p = D.pop(q, None)
        if p is None:
            D[q*q] = q
            yield q
        else:
            x = p + q
            while x in D or not (x&1):
                x += p
            D[x] = p

def get_primes_erat(n):
  return list(itertools.takewhile(lambda p: p<n, erat2()))

st = time.perf_counter()
sum(get_primes_erat(2000000)), time.perf_counter()-st
```
